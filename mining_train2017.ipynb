```python
import pandas as pd

# Đọc file CSV chứa kết quả phân tích diện tích vật thể
csv_file_path = "Project_test2/data2/object_analysis2.csv"
df = pd.read_csv(csv_file_path)

# Tính tỷ lệ xuất hiện của từng nhóm phần trăm diện tích vật thể
area_category_counts = df['area_category'].value_counts(normalize=True) * 100  # Tính theo %
print("Tỷ lệ xuất hiện của từng nhóm phần trăm diện tích vật thể:")
print(area_category_counts)
```

```python
# Thống kê tổng quan về phần trăm diện tích vật thể
print("\nThống kê tổng quan về phần trăm diện tích vật thể:")
print(df['area_percentage'].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Trực quan hóa phân bố các nhóm phần trăm diện tích
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='area_category', order=["0-1%", "1-10%", "10-20%", "20-70%", "70-100%"])
plt.title("Phân bố số lượng vật thể theo nhóm phần trăm diện tích")
plt.xlabel("Nhóm phần trăm diện tích")
plt.ylabel("Số lượng vật thể")
plt.show()
```

```python
# Trực quan hóa biểu đồ tỷ lệ xuất hiện của từng nhóm phần trăm diện tích
plt.figure(figsize=(10, 6))
area_category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Tỷ lệ xuất hiện của từng nhóm phần trăm diện tích vật thể")
plt.ylabel("")  # Bỏ nhãn trục y để biểu đồ trông rõ hơn
plt.show()
```

```python
# Biểu đồ phân phối phần trăm diện tích vật thể
plt.figure(figsize=(10, 6))
sns.histplot(df['area_percentage'], bins=20, kde=True, color='skyblue')
plt.title("Phân phối phần trăm diện tích vật thể")
plt.xlabel("Phần trăm diện tích vật thể (%)")
plt.ylabel("Số lượng vật thể")
plt.show()
```

```python
# Phân bố diện tích vật thể có dính nền hay không
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='background_check', palette="viridis")
plt.title("Phân bố vật thể theo trạng thái dính nền")
plt.xlabel("Trạng thái dính nền")
plt.ylabel("Số lượng vật thể")
plt.show()
```

```python
# Phân bố phần trăm diện tích vật thể theo trạng thái dính nền
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='background_check', y='area_percentage', palette="viridis")
plt.title("Phân bố phần trăm diện tích vật thể theo trạng thái dính nền")
plt.xlabel("Trạng thái dính nền")
plt.ylabel("Phần trăm diện tích vật thể (%)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv("Project_test2/data2/object_analysis2.csv")

# Danh sách các nhóm diện tích
area_categories = ["0-1%", "1-10%", "10-20%", "20-70%", "70-100%"]

# Lọc dữ liệu cho từng nhóm diện tích
filtered_data = {category: df[df['area_category'] == category] for category in area_categories}

# Vẽ biểu đồ phân phối cho từng nhóm diện tích
plt.figure(figsize=(12, 6))

# Duyệt qua các nhóm diện tích và vẽ histogram cho mỗi nhóm
colors = ['red', 'blue', 'green', 'purple', 'orange']  # Màu sắc cho các nhóm diện tích
for i, category in enumerate(area_categories):
    sns.histplot(filtered_data[category]['area_percentage'], kde=True, color=colors[i], label=category)

# Thêm tiêu đề, nhãn trục và legend
plt.title("Phân phối phần trăm diện tích trong các nhóm diện tích")
plt.xlabel("Phần trăm diện tích")
plt.ylabel("Tần suất")
plt.legend(title="Nhóm diện tích")

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("Project_test2/data2/object_analysis2.csv")

# Danh sách các nhóm diện tích
area_categories = ["0-1%", "1-10%", "10-20%", "20-70%", "70-100%"]

# Lọc dữ liệu cho từng nhóm diện tích
filtered_data = {category: df[df['area_category'] == category] for category in area_categories}

# Thống kê chi tiết cho từng nhóm diện tích
for category in area_categories:
    summary = filtered_data[category]['area_percentage'].describe()
    print(f"Thống kê chi tiết cho nhóm diện tích {category}:")
    print(summary)
    print("\n")
```
